```python
import pandas as pd
from glob import glob
```

```python
all_files = glob('../*.csv')

df = pd.concat((pd.read_csv(f) for f in all_files), ignore_index=True)
```

```python
df = df.sort_values(by='starttime')
```

```python
df = df.drop(['birth year', 'gender'], axis=1)
```
